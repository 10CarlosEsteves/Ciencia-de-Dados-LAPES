```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
supermarket_df = pd.read_csv("supermarket_sales - Sheet1.csv")
```

```python
display(supermarket_df)
```

```python
# 1 - Exclua as colunas: "cogs", "gross margin percentage", "gross income".

"""
  Explicação:
  O método drop é utilizado para excluir linhas ou colunas.O primeiro argumento
  do método são as linhas ou colunas e o segundo é axis. axis recebe 0 para
  linha(s) ou 1 para coluna(s). Como extra, tratamos a coluna Date para o
  formato Datetime para manipular a data da melhor forma possível
"""

supermarket_df = supermarket_df.drop(["cogs", "gross margin percentage", "gross income"], axis=1)
supermarket_df['Date'] = pd.to_datetime(supermarket_df['Date'])
```

```python
display(supermarket_df)
```

```python
# 2 - Utilize Regex para identificar IDs com terminações entre '47' e '57' e substitua esses dois últimos valores por '00'.

"""
  Explicação:
  O seguinte trecho de código cria uma lista de id's que possuem a terminação 47
  ou 57. Por padrão, o findall() retorna uma lista vazia caso não encontre o
  padrão regex. Para contornar o problema, filtramos com len(id)>0.

  Após a criação da lista de id's, a lista é percorrida com um loop for, a string
  é trada com dois 0's e depois substituida na linha e coluna correspondente ao
  antigo valor. Nesta seção é exibido os resultados antes e depois para fins de
  comparação.
"""

ids = [id for id in supermarket_df['Invoice ID'].str.findall('[0-9]{3}[-][0-9]{2}[-][0-9]{2}[4|5]7') if len(id)>0]

print("Antes:")
print(ids)

for id in ids:
  id = id[0]
  string_id = list(id)
  string_id[-2],string_id[-1]  = "0", "0"
  string_id = "".join(string_id)
  supermarket_df.loc[supermarket_df['Invoice ID'] == id, 'Invoice ID'] = string_id

print("\nDepois:")
print([id for id in supermarket_df['Invoice ID'].str.findall('[0-9]{3}[-][0-9]{2}[-][0-9]{2}[4|5]7') if len(id)>0])
```

```python
# 3 - Qual é a média mensal de vendas?

"""
  Explicação:
  O código abaixo cria uma indexação com o periodo mensal apartir da coluna
  'Date'. Com essa indexação, é agrupado com a coluna 'Total' e calculada a
  média. Resetamos o index para poder usar a coluna 'Date'.

  Apresentamos o Dataframe com a média de vendas de cada mês. Também apresentamos
  a média de vendas dos 3 meses usados no dataframe.
"""

mm_vendas = supermarket_df.groupby(pd.PeriodIndex(supermarket_df['Date'], freq="M"))['Total'].mean().reset_index()
print("Média de vendas por mês:")
display(mm_vendas)

print(f"Média mensal de vendas: USD {mm_vendas['Total'].mean():0.2f}")
```

```python
# 4 - Qual é a linha de produtos mais bem avaliada? Gere um gráfico de Pizza.

"""
  Explicação:
  O seguinte código gera um gráfico de pizza com a respectiva porcentagem do
  somatório das avaliações de cada linha de produto.

  Junto, é mostrado o top 3 das linhas de produtos mais bem avaliadas que é
  calculada através da média do agrupamento das linhas de produtos com a
  avaliação.
"""

fig = px.pie(
    supermarket_df,
    values='Rating',
    names='Product line',
    title='Linhas de Produtos Avaliadas'
)
fig.show()

print("Top 3 das linhas de produtos bem avaliados")
valued = supermarket_df.groupby('Product line')['Rating'].mean().reset_index().sort_values(by='Rating', ascending=False)
display(valued.head(3))
```

```python
# 5 - Quais são os 5 produtos mais caros de acordo com a linha de produtos?

"""
  Explicação:
  O código abaixo extrai as ocorrências únicas das linhas de produtos. Com isso,
  temos uma lista com todas as linhas de produtos disponíveis. Usamos um loop
  for para percorrer as linhas de produtos, extraimos da tabela as linhas em que
  contem a linha de produto atual, depois ordenamos em ordem descrescente e
  pegamos os 5 primeiros valorers com o método head().
"""

product_lines = supermarket_df['Product line'].unique()

for product_line in product_lines:
  cinco_mais_caros = supermarket_df.loc[supermarket_df['Product line'] == product_line]
  cinco_mais_caros = cinco_mais_caros.sort_values(by='Unit price', ascending=False).head()
  display(cinco_mais_caros)
```

```python
# 6 - Quais são os 10 produtos mais populares (vendidos com maior frequência)?
# E os 10 menos populares?

"""
  Explicação:
  O código abaixo ordena os valores pela quantidade vendida em ordem crescente e
  decrescente e salva em duas variaveis. Os valores retornados são Dataframes,
  logo os Dataframes são exibidos com o método display().
"""

mais_populares = supermarket_df.sort_values(by='Quantity', ascending=False)[:10]
menos_populares = supermarket_df.sort_values(by='Quantity', ascending=True)[:10]

print("10 PRODUTOS MAIS POPULARES")
display(mais_populares)
print("10 PRODUTOS MENOS POPULARES")
display(menos_populares)
```

```python
# 7 - Existe alguma correlação entre o preço de um produto e a quantidade vendida?

"""
  Explicação:
  O código abaixo mostra a correlação entre o preço de um produto e a quantidade
  vendida através do método Pearson. Realizamos uma cópia profunda para não
  ocorrer nenhuma modificação acidental no Dataframe original. Após a cópia, é
  realizado a correlação de Pearson através do método pandas corr(). O resultado
  obtido é uma matriz de correlações.

  O método de correlação de Pearson é um método que permite verificar se duas
  variáveis estão relacionadas entre si. o método varia de -1 a 1, com -1 sendo
  duas variáveis inversamente proporcionais, 1 sendo duas variáveis proporcionais
  e quando os números forem próximos de 0 indicam baixa correlação entre si.

  Os resultados obtidos são mostrados em forma de gráfico, Dataframe e uma frase.
  O gráfico de disperção associa os preços únicos com a quantidade. A matriz de
  correlações é mostrada junto com o resultado obtido da correlação.
"""

preco_produto = supermarket_df[['Unit price', 'Quantity']].copy(deep=True)
matriz_relacao = preco_produto.corr(method='pearson')
correlacao = matriz_relacao.loc['Quantity', 'Unit price']

fig = px.scatter(supermarket_df,x=supermarket_df['Unit price'], y=supermarket_df['Quantity'])
fig.show()

display(matriz_relacao)
print(f"A correlação obtida entre a preço de um produto e quantidade é de {correlacao:0.3f} no método de Pearson.")
```

```python
# 8 - Existe alguma correlação entre o preço de um produto e o método de pagamento? Gere um gráfico de Dispersão.

"""
  Explicação:
  O código abaixo gera um gráfico de dispersão com as relaçãoes entre método de
  pagamento e o preço único.
"""

fig = px.scatter(supermarket_df,
                 x=supermarket_df['Payment'],
                 y=supermarket_df['Unit price'],
                 color=supermarket_df['Product line'],
                 symbol=supermarket_df['Product line']
                 )
fig.show()
```

```python
# 9 - Qual é o perfil de cliente que mais compra? Leve em conta cidade, tipo de cliente e gênero.

"""
  Explicação:
  O código abaixo gera um gráfico de linhas com o perfil de clientes que mais
  realizam compras. Para traçar o perfil, é levado em consideração a cidade,
  tipo de cliente e seu gênero.

  Para construir o gráfico, é agrupado a cidade, tipo de cliente e gênero e é
  somado o Total gasto. O resultado é um mini Dataframe em que cada combinação
  de cidade, tipo e gênero é uma linha diferente. Para plotar as multiplas
  combinações como barras únicas, o eixo x é considerado como o index da linha e
  no método update_layout é formatado o texto que deve aparecer para representar
  o perfil de clientes.

  O resultado final é um gráfico do perfil de cliente por gasto em que cada
  perfil de cliente corresponde a uma barra no gráfico de barras.
"""

perfil_cliente = supermarket_df.groupby(['City', 'Customer type', 'Gender'])['Total'].sum().reset_index()

fig = px.bar(perfil_cliente, x=perfil_cliente.index, y='Total', color='Total')

fig.update_layout(
    xaxis = dict(
        tickmode = 'array',
        tickvals = perfil_cliente.index,
        ticktext = [ f"{row['City']}-{row['Customer type']}-{row['Gender']}" for index, row in perfil_cliente.iterrows()]
    )
)
fig.show()
```

```python
# 10 - Quais são os meses de pico de vendas? E os meses mais fracos? Gere um gráfico de sua escolha para exemplificar essas diferenças.

"""
  Explicação:
  O código abaixo pega uma amostra do Dataframe original e copia para evitar
  modificação no original. Após a cópia, a amostra é ordenada conforme os meses
  e depois é acrescentado a coluna 'Month' a amostra. Após essas operações, é
  somado a quantidade de receita total em cada mês e depois é plotado um gráfico
  do mês e do total de vendas.

  O gráfico permite observar a progressão do faturamento mensal das filiais de
  supermercado. Através deste gráfico, é possível observar os meses de picos
  máximos e mínimos nas arrecadações.
"""

mes_vendas = supermarket_df[['Date', 'Total']].copy(deep=True)
mes_vendas = mes_vendas.sort_values(by='Date')
mes_vendas['Month'] = mes_vendas['Date'].dt.month_name()

relatorio_mensal = mes_vendas.groupby('Month')['Total'].sum().reset_index()

fig = px.line(relatorio_mensal, x='Month', y='Total')

fig.update_layout(
    title="Relatório de Vendas Mensais",
    xaxis_title="Meses",
    yaxis_title="Total (USD)",
)
```

```python
# 11 - Qual filial gera mais receita?

"""
  Explicação:
  O código abaixo agrupa as cidades e soma a coluna Total. Assim, é possível ter
  um Dataframe amostra em que cada filial de uma cidade está associada uma
  receita.

  Através da amostra, extraimos a filial que mais gera receita e sua respectiva
  receita. Após a extração, mostramos uma frase indicando nome da filial e sua
  receita.
"""

receita_filiais = supermarket_df.groupby('City')['Total'].sum().reset_index()
filial_n1, receita_n1 = receita_filiais.max()[0], receita_filiais.max()[1]

print(f"A filial que mais gerou receita foi a filial de {filial_n1} com a receita total de USD$ {receita_n1:0.2f}")
```

```python
# 12 - Há uma diferença no padrão de vendas durante os dias da semana e os fins de semana?

"""
  Explicação:
  O código abaixo extrai as colunas: linhas de produto, preço unico e data do
  Dataframe original e realiza uma cópia para evitar qualquer alteração
  indesejada no Dataframe original.

  Na amostra, é acrescentada a coluna 'Day of Week' através da coluna 'Date' com
  o método day_name(). Depois, é criado uma lista com os dias da semana ordenados
  e ordenamos a coluna 'Day of Week'.

  Por fim, agrupamos a coluna dos dias da semana e somamos a quantidade de
  produtos comprados. O resultado será um mini Dataframe que relaciona cada dia
  da semana uma determinada quantidade de produtos comprados. Com esse mini
  Dataframe, plotamos um gráfico de barras para melhor visualização de padrões.
"""


amostra = supermarket_df[['Product line', 'Quantity', 'Unit price', 'Date']].copy(deep=True)
amostra['Day of Week'] = amostra['Date'].dt.day_name()

ordem = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
amostra['Day of Week'] = pd.Categorical(amostra['Day of Week'], categories=ordem, ordered=True)
semana_vendas = amostra.groupby('Day of Week')['Quantity'].sum().reset_index()

fig = px.bar(semana_vendas, x='Day of Week', y='Quantity', color='Quantity', height=600)
fig.show()
```
